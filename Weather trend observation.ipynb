```python
import pandas as pd

#import dataset & index the year

temperature_compiled = pd.read_csv('results.csv',index_col = 'year',parse_dates = True)
temperature_compiled = temperature_compiled.loc['1849-01-01':'1889-01-01']
```

```python
#take out the na from helsinki temperature
temperature_compiled['hel_temp'].dropna()
```

```python
temperature_compiled['global_temp'].plot(figsize= (30,10))
```

```python
temperature_compiled['mov_heltemp'] = temperature_compiled['hel_temp'].rolling(50).mean()
temperature_compiled['mov_heltemp'].dropna()
#plot helsinki temperature with 50-year moving average
temperature_compiled['mov_heltemp'].dropna().plot(figsize=(30, 10))
```

```python
#plot helsinki temperature without moving average
temperature_compiled['hel_temp'].dropna().plot(figsize=(30, 10))
```
